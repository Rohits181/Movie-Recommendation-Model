```python
clean_set = movies_set.dropna(subset = ["overview"])
```

```python
clean_set.isnull().sum()
```

```python
new_set = clean_set[['original_title',"overview","popularity","vote_average","vote_count"]]
```

```python
new_set
```

```python
import tensorflow as tf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
vectorizer = TfidfVectorizer(min_df=3, ngram_range=(1, 3), analyzer='word', stop_words='english')
tf_matrix = vectorizer.fit_transform(new_set['overview'])
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

corr = sigmoid_kernel(tf_matrix,tf_matrix)
```

```python
corr
```

```python

indices = pd.Series(new_set.index,index=new_set['original_title']).drop_duplicates()
```

```python
indices
```

```python

def recommend(name,sig=corr):
    idx = indices[name]
    movie_scores = list(enumerate(sig[idx]))
    movie_scores = sorted(movie_scores,key = lambda x:x[1],reverse=True)
    top_10_similar_movies_indices = movie_scores[1:11]
    movies_indices = [i[0] for i in top_10_similar_movies_indices]
    return new_set['original_title'].iloc[movies_indices]
```

```python
recommendations = recommend('Winnie Mandela')
```
